```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score
import rasterio
from rasterio.plot import show
from rasterio.mask import mask
from matplotlib import pyplot as plt
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, matthews_corrcoef, confusion_matrix, classification_report
from rasterio import plot
import numpy as np
from sklearn.svm import SVC
import os
import rioxarray
import json
from time import time
import seaborn as sns
import pandas as pd
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
def get_building_labels(building_footprint_poly, raster_data):
    ## Load Buliding Footprints
    with open(building_footprint_poly) as building_poly:
        data = json.load(building_poly)
    ## Get the Spatial References of Building Footprint Polygons (Geographic Projection)
    spatial_ref = data["crs"]["properties"]["name"]
    geometries = [feat["geometry"] for feat in data["features"]]
    ## Define an Empty A
    rds = rioxarray.open_rasterio(raster_data).isel(band=0)
    rds.values[:] = 1
    rds.rio.write_nodata(0, inplace=True)
    # Tagg All Pixels Inside Building Footprints as 1
    building_categories = rds.rio.clip(geometries, spatial_ref, drop=False)
    
    return building_categories
```

```python
def clean_satellite_image(raster_data, viz_scheme=None):
    dataset = rasterio.open(raster_data)
    if viz_scheme:
        image = dataset.read(raster_viz_schemes[viz_scheme])
    else:
        image = dataset.read()
            
    image = np.interp(image,(np.min(image), np.max(image)), (0, 255)).astype(np.uint8)
    return image
```

```python
def vectorize_raster(raster_loc):
    dataset = rasterio.open(raster_loc)
    image = dataset.read()
    dim_raster = image.shape
    image = np.reshape(image, (dim_raster[0], dim_raster[1]*dim_raster[2])).T
    
    return image
```

```python
def split_train_test_index(num_sample, perc_train):
    ## Define the number of training samples
    num_train = int(np.ceil(num_sample * perc_train/100))
    ## Define a randomized index list for sampling
    index_list = np.random.permutation(num_sample)
    
    train_index = index_list[:num_train]
    test_index = index_list[num_train:]
    
    return train_index, test_index
```

```python
def vectorize_target(target_mat):
    arr = target_mat.values
    return arr.reshape(arr.shape[0]*arr.shape[1])
```

```python
def create_training_data(raster_folder, footprint_folder, train_id):
    ## Define Folders for Raster and Footprint Data
    raster_list = os.listdir(raster_folder)
    footprint_list = os.listdir(footprint_folder)
    ## Define Data Read Locations
    raster_loc = os.path.join(raster_folder, raster_list[train_id[0]])
    footprint_loc = os.path.join(footprint_folder, footprint_list[train_id[0]])
    ## Get the First Samples
    predictor_vars = vectorize_raster(raster_loc)
    target_var = get_building_labels(footprint_loc,raster_loc)
    target_var = vectorize_target(target_var)
    
    for get_ind in train_id[1:]:
        ## Define Data Read Locations
        try:
            raster_loc = os.path.join(raster_folder, raster_list[get_ind])
            footprint_loc = os.path.join(footprint_folder, footprint_list[get_ind])
            print(raster_loc)
            ## Read in the Next Raster and Footprint Data
            read_raster = vectorize_raster(raster_loc)
            read_target = get_building_labels(footprint_loc,raster_loc)
            read_target = vectorize_target(read_target)
            ## Append the New Data to Main Arrays
            predictor_vars = np.append(predictor_vars, read_raster, axis=0)
            target_var = np.append(target_var, read_target, axis=0)
        except:
            pass
    return predictor_vars, target_var
    
```

```python
raster_viz_schemes = {'Natural Color':[4,3,2],
                      'Color (urban)':[7,6,4],
                      'Color Infrared (vegetation)':[5,4,3],
                      'Agriculture':[6, 5, 2],
                      'Atmospheric Penetration':[7,6,5],
                      'Healthy Vegetation':[5,6,2],
                      'Land/Water':[5,6,4],
                      'Natural With Atmospheric Removal':[7,5,3],
                      'Shortwave Infrared':[7,5,4],
                      'Vegetation Analysis':[6,5,4]}
```

```python
num_sample = len(raster_list)
perc_train = 70
train_ind, test_index = split_train_test_index(num_sample, perc_train)
X_train, Y_train = create_training_data(raster_folder, footprint_folder, train_ind)
X_test, Y_test = create_training_data(raster_folder, footprint_folder, test_index[1:])
```

## ESDA

```python
## Remove Dark Spots of Satellite Data (All Bands ==0)
dark_spots_train = np.where(X_train.sum(axis=1)!=0)
dark_spots_test = np.where(X_test.sum(axis=1)!=0)

## Remove Dark Spot Pixels from Training and Test
Y_train = Y_train[dark_spots_train]
X_train = X_train[dark_spots_train]

X_test = X_test[dark_spots_test]
Y_test = Y_test[dark_spots_test]
```

```python
X = np.vstack((X_train,X_test))
Y = np.hstack((Y_train, Y_test)).T

df = pd.DataFrame(columns = ['Band 1 Visible','Band 2 Visible',
                             'Band 3 Visible','Band 4 Red',
                             'Band 5 Near-Infrared','Band 6 SWIR',
                             'Band 7 SWIR','Band 8 Panchromatic'])


for ind, col in enumerate(df.columns):
    df[col] = X[:,ind]

df['Building Label'] = Y
```

```python
sns.set(font_scale=2)
plt.figure(figsize=(20,10))
plt.gcf().subplots_adjust(bottom=0.30)
corr = df[:-1].corr(method = 'spearman')

sns.heatmap(corr, annot=False)

plt.savefig('figs/space_net_spearman_mat.png', dpi=300)
```

```python
fig, ax = plt.subplots(figsize=(20,10))
sns.set_style("ticks")
plt.gcf().subplots_adjust(bottom=0.30)
df_plot = pd.melt(df, value_vars=df.columns[:-1], id_vars=df.columns[-1])
ax = sns.violinplot(x='variable', y='value', scale='width', ax=ax, split = True, hue =df.columns[-1], data=df_plot)
plt.setp(ax.get_xticklabels(), rotation=45)
ax.set(xlabel=None)
ax.set(ylabel=None)
plt.savefig('figs/space_net_violin.png', dpi=300)
```

### Fit Decision Tree Classifier

```python
t_tree = time()
dtree = DecisionTreeClassifier(criterion='gini', random_state = 0)
clf = dtree.fit(X_train, Y_train)
t_tree = time() - t_tree
t_tree
```

```python
print('The default tree depth is {0}'.format(clf.get_depth()))
print('In Sample (Training) Accuracy : {0}'.format(accuracy_score(Y_train, clf.predict(X_train))))
print('Out of Sample (Test) Accuracy : {0}'.format(accuracy_score(Y_test, clf.predict(X_test))))
```

```python
max_depth = clf.get_depth()

in_sample = []
out_sample = []
for depth in range(max_depth, 2, -1):
    dtree = DecisionTreeClassifier(criterion='entropy', max_depth=depth, random_state = 55)
    clf_trial = dtree.fit(X_train, Y_train)
    in_sample.append(accuracy_score(Y_train, clf_trial.predict(X_train)))
    out_sample.append(accuracy_score(Y_test, clf_trial.predict(X_test)))
    
accuracy_df = pd.DataFrame()
accuracy_df['In Sample (Training)'] = in_sample
accuracy_df['Out Sample (Test)'] = out_sample
accuracy_df['Tree Depth'] = list(range(max_depth, 2, -1))
f, ax = plt.subplots(1, 1)
ax = sns.lineplot(x='Tree Depth', y= 'In Sample (Training)', data = accuracy_df, legend= "full")
ax = sns.lineplot(x='Tree Depth', y= 'Out Sample (Test)', data = accuracy_df, legend= "full")
plt.show()
```

```python
params = {'max_depth': np.linspace(2,30, 29)}

grid_search_cv = GridSearchCV(DecisionTreeClassifier(random_state=55), 
                              params, verbose=1, cv=3)

grid_search_cv.fit(X_train, Y_train)
```

```python
dtree_best = grid_search_cv.best_estimator_
clf_best = dtree_best.fit(X_train, Y_train)
print('In Sample (Training) Accuracy : {0}'.format(accuracy_score(Y_train, clf_best.predict(X_train))))
print('Out of Sample (Test) Accuracy : {0}'.format(accuracy_score(Y_test, clf_best.predict(X_test))))
```
