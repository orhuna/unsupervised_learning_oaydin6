```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib as mpl
import numpy as np
import os
import pandas as pd
from time import time
import seaborn as sns
from sklearn import tree
from matplotlib import rcParams
from yellowbrick.model_selection import LearningCurve
from sklearn import preprocessing
```

```python
sns.set_theme(style="whitegrid")
seagrass_data_loc = r'data/Seagrass_data'
seagrass = pd.read_csv(os.path.join(seagrass_data_loc,'presence.txt'), sep='\t')
## Wrangle Temporal Data and Use Summaries
## Drop ID field
seagrass = seagrass.drop(columns = 'ID')
seagrass = seagrass.drop(columns = 'LONGITUDE')
seagrass = seagrass.drop(columns = 'LATITUDE')
## Drop monthly measurements
temporalVars = [var for var in seagrass.columns if '-01' in var]
seagrass = seagrass.drop(columns = temporalVars)

## Drop 
seasonalVars = [var for var in seagrass.columns if '2015' in var and 'year' not in var]
seagrass = seagrass.drop(columns = seasonalVars)
seagrass.head()

## Extreme Vars 
extremeVars = [var for var in seagrass.columns if 'max' in var or 'min' in var]
seagrass = seagrass.drop(columns = extremeVars)
seagrass.head()

## Make Columns Human-Readable

seagrass = seagrass.rename(columns={'BIO_CLASS':'Species Name','VOTEMPER_2015_year':'Temperature',
            'VOSALINE_2015_year':'Salinity', 'CHL_2015_year':'Chlorophyll',
            'NIT_2015_year':'Nitrate', 'ZSD_2015_year':'Secchi Disk Depth',
            'PHO_2015_year':'Phosphate', 'VHM0_2015_year':'Wave Height',
            'Med_bathym':'Water Depth', 'Distance_to_Major_Cities':'Dist. to Major City',
            'Distance_to_Complete_Cities': 'Dist. to City','Distance_to_Port':'Dist. to Port',
            'Distance_to_Major_River': 'Dist. Major River', 'Distance_to_Complete_River': 'Dist. to River',
            'Distance_to_Coast':'Dist. to Coast'})

## Drop Water Depth More than 90 m (Seagrasses cannot live past 90 m depth)
seagrass.drop(seagrass[seagrass['Water Depth']<-90].index, inplace=True)
seagrass.head()
X_seagrass = seagrass[seagrass.columns[1:-1]]
Y_seagrass = seagrass[seagrass.columns[0]]
```

```python
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score, calinski_harabasz_score
from yellowbrick.cluster import KElbowVisualizer, InterclusterDistance
from time import time

class EM_label(GaussianMixture):
    def __init__(self, n_clusters=1, **kwargs):
        ## Modify the n_components for search with YB
        kwargs["n_components"] = n_clusters
        super(EM_label, self).__init__(**kwargs)

    def fit(self, X):
        ## Produce Most Likely Clusters
        super(EM_label, self).fit(X)
        self.labels_ = self.predict(X)
        return self

class run_clustering(object):
    def __init__(self, model, data, max_k, cov_type = 'full', n_init=10, max_iter=300, tol=1e-4, random_state = 12345, 
                 reduction_alg = None, plt_folder=None, exp_name=None):
        
        self.model = model
        self.data = data
        self.max_k = max_k
        self.n_init = n_init
        self.max_iter = max_iter
        self.tol = tol
        self.random_state = random_state
        self.reduction_alg = reduction_alg
        self.plt_folder = plt_folder
        
        if self.max_k < 2:
            ## Make sure to set the righ number of clusters here
            raise SystemExit('Number of Clusters cannot be less than two')
            
        ## Define Output Objects
        self.k = []
        self.run_times = []
        self.n_iters = []
        self.model_def = []
        self.labels = []
        
        if model.upper() == 'KMEANS':
            self.cluster_centers = []
            self.inertia = []
            self.silhouette = []
            self.ch_index = []
            
        if model.upper() == 'EM':
            self.bic = []
            self.converged = []
            self.means = []
            self.weights = []
            self.precision = []
        
    def search_k_means(self):
        for k in range(2, self.max_k + 1):
            print('Solving for {0} clusters'.format(k))
            self.k.append(k)
            t_tot = time()
            kmeansRun = KMeans(n_clusters=k, n_init=self.n_init, max_iter=self.max_iter, 
                            tol=self.tol, random_state = self.random_state).fit(self.data)
            t_tot = time() - t_tot
            print('{0} clusters solved in {1} seconds'.format(k, t_tot))
            self.run_times.append(t_tot)
            self.cluster_centers.append(kmeansRun.cluster_centers_)
            self.labels.append(kmeansRun.labels_)
            self.inertia.append(kmeansRun.inertia_)
            self.n_iters.append(kmeansRun.n_iter_)
            self.model_def.append(kmeansRun)
            #self.silhouette.append(silhouette_score(self.data, kmeansRun.labels_))
            #self.ch_index.append(calinski_harabasz_score(self.data, kmeansRun.labels_))
            print('Done')
            
    def search_EM(self):
        for k in range(2, self.max_k + 1):
            self.k.append(k)
            t_tot = time()
            EM = GaussianMixture(n_components=k, random_state = self.random_state).fit(self.data)
            t_tot = time() - t_tot
            self.run_times.append(t_tot)
            self.weights.append(EM.weights_)
            self.means.append(EM.means_)
            self.bic.append(EM.bic(self.data))
            self.n_iters.append(EM.n_iter_)
            self.precision.append(EM.precisions_)
            self.model_def.append(EM)
            self.labels.append(EM.predict(self.data))
```

```python
def viz_labels_matrix(X_df, labels, loc):
        X = X_df.copy()
        X['Labels'] = labels
        sns.pairplot(X, hue='Labels', palette = 'bright')
        if loc:
            plt.savefig(loc)
            plt.close()
```

```python
from sklearn.decomposition import FastICA
def search_ica(X, max_iter=200, loc=None, random_seed = 12455):
    kurtosis = []
    inds = []
    for k in range(2, X.shape[1]+1):
        ica = FastICA(n_components=k, max_iter=max_iter, random_state=random_seed)
        ica_fitted = ica.fit_transform(X)
        ica_fitted_df = pd.DataFrame(ica_fitted)
        kurtosis.append(ica_fitted_df.kurtosis().abs().mean())
        inds.append(k)
    if loc:
        plt.plot(inds, kurtosis)
        plt.xlabel('Number of Components')
        plt.ylabel('Mean of Absolute Kurtosis')
        plt.xticks(inds)
        plt.savefig(loc)
```

```python
from sklearn import random_projection
from sklearn.metrics import mean_squared_error

def random_projection_profile(X, n_permutation=1):
    err = np.zeros((n_permutation, X.shape[1]))
    k_list = []
    for k in range(0, X.shape[1]):
        for perm in range(n_permutation):
            transformer = random_projection.GaussianRandomProjection(n_components=k+1)
            X_new = transformer.fit_transform(X)
            approx_inv = np.linalg.pinv(transformer.components_.T)
            recreate = X_new.dot(approx_inv)
            err[perm, k] = mean_squared_error(X, recreate)  
        k_list.append(k+1)
    return err, k_list
```

```python
err, k = random_projection_profile(X_spacenet, n_permutation=100)
```

```python
plt.errorbar(k, list(err.mean(axis=0)), list(np.ptp(err,axis=0)/2))
plt.xlabel('Number of Components')
plt.ylabel('Reconstruction Error (MSE)')
plt.savefig('figs/Q2/random_proj_spacenet_num_comp.png')
```

```python
from yellowbrick.model_selection import LearningCurve
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import LabelEncoder
from sklearn.neural_network import MLPClassifier as NN

# Encode the Prediction Variable
y = LabelEncoder().fit_transform(Y_spacenet)
# Create the learning curve levels
cv = StratifiedKFold(n_splits=12)
sizes = np.linspace(0.3, 1.0, 10)
# Define Neural Network Classifier
model = NN(hidden_layer_sizes = (4,4), activation = 'tanh', alpha=0.0001)

df_list = [X_spacenet_kpca_df, X_spacenet_pca_df, X_spacenet_rp_df, X_spacenet_ica_df]
method_list = ['kpca', 'pca', 'rp', 'ica']
```
